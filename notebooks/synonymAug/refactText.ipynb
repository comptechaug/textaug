```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
# vim:fileencoding=utf-8

import pandas as pd
import re

with open("Sinonims.txt") as sFile:
	strWithAllFile = sFile.read()

#deliting usles symbols like '.'
pattern = r'См.|\.|см.|ср.|Ср|;'
strWithAllFile = re.sub(pattern,'',strWithAllFile)

#deliting all in brackets [],(),<>,'',""
#including them
pattern = r'\[(.)*\]|\<(.)*\>|\((.)*\)|\'(.)*\'|\"(.)*\"'
strWithAllFile = re.sub(pattern,'',strWithAllFile)
strWithAllFile = re.sub(r'--(.)*,',',',strWithAllFile)
strWithAllFile = re.sub(r'#,|,#','#',strWithAllFile)
strWithAllFile = re.sub(r'\|\|(.)*\n','\n',strWithAllFile)
```

```python
def changeChar2Str(text,placeOfChar,str):
	return text[:placeOfChar] + str + text[placeOfChar + 1:]

def placeOfBeginWord(text,indexOfGrid):
	i = 0
	while text[indexOfGrid - i] != "\n":
		i+=1
		if(i > 100):
			print("someThing Wrong With Text")
			print("index with errror:",indexOfGrid)
			break
	return indexOfGrid - i

def replaceGridToBeginWord(text,beg,end):
	newText = changeChar2Str(text,end,",")
	newText = changeChar2Str(newText,beg,"\n#")
	return newText

def replaceAllGridtoBeginOfWords(text):
	newtext = "  " + text
	i = 2
	while i < len(newtext):
		if newtext[i] == "#":
			beginOfwrd = placeOfBeginWord(newtext,i)
			endOfwrd   = i
			newtext    = replaceGridToBeginWord(newtext,beginOfwrd,endOfwrd)
			i+=1
		i+=1
	return newtext
```

```python
factorSetOfWords = strWithAllFile.split("#")
factorSetOfWords[0:5]
```

```python
for i in range(len(factorSetOfWords)):
    string = factorSetOfWords[i]
    #erase \n 
    string = re.sub(r'\n','',string)
    # "__" to "_"
    string = re.sub(r'  ',' ',string)
    # ",_" to "," 
    string = re.sub(r', ',',',string)
    
    string = string.lower()
    string = string.split(",")
    
    ##here need erasing wrds like "adad@fdsg"
    #
    #
    #######################################
    
    factorSetOfWords[i] = string
```

```python
factorSetOfWords[0][0] = "а"
for lists in factorSetOfWords:
    index = 0
    for word in lists:
        if len(word) > 21 and index != 0:
            print(lists.pop(index))
        index +=1
print(len(factorSetOfWords[0]))
factorSetOfWords[:10]
```
